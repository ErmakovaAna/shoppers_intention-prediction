## Общее исследование данных

- ```Administrative```, ```Informational```, ```ProductRelated``` –  количество страниц указанного типа, которое посетил пользователь

- ```Administrative_Duration```, ```Informational_Duration```, ```ProductRelated_Duration``` – количество времени, проведенное на странице указанной категории

- ```BounceRates``` (показатель отказов) – процент сессий, во время которых была просмотрена *только* определенная страница, от общего количества сессий, начавшихся с просмотра этой страницы

- ```ExitRates``` (показатель выходов) – процент сессий, которые заканчиваются просмотром данной страницы

- ```PageValues``` – средняя ценность страниц, рассчитываемая путем деления общего дохода на количество сессий, в рамках которых посещалась данные страницы

- ```SpecialDay``` – показатель близости времени посещения сайта к некоторому праздничному дню

- ```Month``` – месяц выполнения сессии

- ```OperatingSystems``` – операционная система пользователя

- ```Browser``` – браузер, через который выполнялась сессия

- ```Region``` – регион пользователя

- ```TrafficType``` – категория трафика

- ```VisitorType``` – категория пользователя

- ```Weekend``` – флаг, указывающий на то, выполнялась ли сессия на выходных

- ```Revenue``` – флаг, указывающий на то, была ли завершена покупка в рамках данной сессии

Для проведения экспериментов по сравнению производительности библиотек Pandas и Polars, создаем большой DataFrame путем повторения наших данных 100 раз по вертикальной оси:

````python
duplicated_df = pd.concat([df] * 100, ignore_index=True)
df.shape, duplicated_df.shape
````

### Загрузка данных

````python
%%time

df_pd = pd.read_csv('online_shoppers_big.csv')
````

````python
%%time

df_pl = pl.read_csv('online_shoppers_big.csv')
````

Polars выполняет загрузку данных за 984 ms и показывает лучшую производительность, чем Pandas, у которого эта операция занимает 1.71 s.

### Выбор и фильтрация данных

````python
%%time

df_pd[['PageValues', 'Revenue']]
````

````python
%%time

df_pl[['PageValues', 'Revenue']]
````

Polars в данном эксперименте почти в 70 раз быстрее, чем Pandas, выполняет выбор данных из датасета: 96.1 µs (или 0.0961 ms) против 6.7 ms.

````python
%%time

df_pd.query(
    'VisitorType == "New_Visitor" and Revenue == True and ProductRelated_Duration < 10'
)
````

````python
%%time

df_pl.filter(
    (pl.col('VisitorType') == 'New_Visitor') &
    (pl.col('Revenue') == True) &
    (pl.col('ProductRelated_Duration') < 10)
)
````

При фильтрации данных Polars также показывает лучшую производительность по сравнению с Pandas. Pandas выполнил  фильтрацию за 57.8 ms, а Polars справился почти в 2 раза быстрее, за 28 ms.

### Агрегация данных

````python
%%time

df_pd.groupby('VisitorType').agg(
    {
        'PageValues': 'mean',
        'Weekend': 'mean',
        'Revenue': 'mean'
    }
).sort_values(by='Revenue', ascending=False)
````

````python
%%time

df_pl.groupby('VisitorType').agg(
    [
        pl.mean('PageValues'),
        pl.mean('Weekend'),
        pl.mean('Revenue')
    ]
).sort('Revenue', descending=True)
````

````python
%%time

query = (
    df_pl
    .lazy()
    .groupby(by='VisitorType')
    .agg(
        [
           pl.mean('PageValues'),
           pl.mean('Weekend'),
           pl.mean('Revenue')
        ]
    ).sort('Revenue', descending=True)
)
````

````python
%%time

query.collect()
````

Polars несколько быстрее выполнил команды группировки, агрегации и сортировки данных, чем Pandas. Pandas выполнил данные команды за 110 ms, а Polars – за 81 ms.

Использование lazy operations позволило дополнительно уменьшить время на операции группировки, агрегации и сортировки. В сумме выполнение этих команд заняло у Polars 63.213 ms.

*Общий вывод*. Сравнение производительности библиотек Pandas и Polars показало, что в целом Polars работает быстрее, особенно при выполнении операций загрузки, выбора и фильтрации данных. При агрегации данных преимещество Polars заключается в возможности применять отложенные lazy operations, которые показывают более высокую производительность.
